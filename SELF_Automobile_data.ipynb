```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes=True)
%matplotlib inline
```

```python
a = pd.read_csv("Automobile.csv")
```

```python
a.head(11)
```

```python
sns.distplot(a['engine_size'])
```

```python
sns.distplot(a['engine_size'], kde=False, rug=True)
plt.show()
```

```python
j = sns.jointplot(a['engine_size'], a['horsepower'])
import scipy.stats as stats
j.annotate(stats.pearsonr)
plt.pyplot.show()
```

```python
sns.jointplot(a['engine_size'], a['horsepower'], kind="hex")
```

```python
sns.pairplot(a[['engine_size', 'horsepower', 'normalized_losses']])
```

```python
sns.stripplot(a['aspiration'], a['engine_size'], jitter=True)
```

```python
sns.swarmplot(a['aspiration'], a['engine_size'])
```

```python
sns.boxplot(y='horsepower', x='number_of_doors', data=a)
```

```python
sns.boxplot(a['horsepower'], a['number_of_doors'], hue=a['fuel_type'])
```

```python
sns.barplot(a['body_style'], a['horsepower'], hue=a['engine_location'])
```

```python
sns.countplot(a['body_style'], hue=a['engine_location'])
```

```python
sns.pointplot(a['body_style'], a['horsepower'])
```

```python
sns.factorplot(x="fuel_type", y="horsepower", data=a, kind="violin", hue='number_of_doors', row='engine_location')
#point, bar, count, box, violin, strip
```

```python
sns.lmplot(x='horsepower', y='peak_rpm', data=a, hue='fuel_type')
```
