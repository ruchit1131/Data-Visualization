```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes=True)
%matplotlib inline
```

```python
games = pd.read_csv('games.csv')
games.info()
```

```python
games.dropna(inplace=True)
```

```python
games.info()
```

```python
sns.distplot(games['average_rating'], kde=False)
```

```python
sns.jointplot(games['minage'], games['average_rating'], kind='reg')
```

```python
sns.pairplot(games[['minage','average_rating','playingtime']])
```

```python
sns.stripplot(games['type'], games['playingtime'], jitter=True)
```

```python
sns.regplot(x='playingtime', y='average_rating', data=games[games['playingtime'] <500])
```
